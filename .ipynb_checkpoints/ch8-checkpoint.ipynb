```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font_path = 'NanumGothic-Regular.ttf'
fontprop = fm.FontProperties(fname=font_path, size=18)

plt.rc('font', family='Nanum Gothic')
month = ['mar', 'apr', 'may', 'jun', 'jul']
sales = [1, 5, 7, 3, 7]
plt.title('한글', fontproperties=fontprop)
plt.plot(month, sales, color='y')
```

```python
plt.figure(dpi=150)
plt.title('font', family='Nanum Gothic')
plt.title('월별 판매 실적')
plt.plot(month, sales, color='b', label='단위: 개수')
plt.legend(loc='lower right')
plt.xlabel('월')
plt.ylabel('매출')
```

```python
import random

height, weight = [], []
for i in range(100):
    height.append(random.randint(100, 200))
    weight.append(random.randint(20, 100))
plt.rc('font', family='Nanum Gothic')
plt.title('키와 몸무게의 상관관계')
plt.scatter(height, weight, c=height, cmap='RdPu')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
```

```python
import random

height, weight, score = [], [], []
for i in range(100):
    height.append(random.randint(100, 200))
    weight.append(random.randint(20, 100))
    score.append(random.randint(1, 1000))
plt.rc('font', family='Nanum Gothic')
plt.title('키와 몸무게의 상관관계')
plt.scatter(height, weight, s=score, c=score, cmap='rainbow', alpha=0.7)
plt.xlabel('키')
plt.ylabel('몸무게')
plt.colorbar(label='체육 점수')
plt.show()
```

```python
single = ['a', 'b', 'c', 'd', 'e']
week1 = [42, 58, 19, 92, 84]
week2 = [53, 52, 48, 98, 73]

plt.rc('font', family='Nanum Gothic')
plt.title('오디션 프로그램 득표 현황')
plt.plot(single, week1, label='첫째 주', color='hotpink')
plt.plot(single, week2, label='둘째 주', color='royalblue')
plt.legend()
plt.show()
```

```python
for i in range(len(week1)):
    week1[i] = -week1[i]

plt.rc('font', family='Nanum Gothic')
plt.title('오디션 프로그램 득표 현황')
plt.barh(single, week1, label='첫째 주', color='hotpink')
plt.barh(single, week2, label='둘째 주', color='royalblue')
plt.legend()
plt.show()
```

```python
import random

height1, weight1, score1, height2, weight2, score2 = [], [], [], [], [], []
for i in range(100):
    height1.append(random.randint(140, 180))
    weight1.append(random.randint(40, 80))
    score1.append(random.randint(0, 200))
    height2.append(random.randint(160, 200))
    weight2.append(random.randint(50, 100))
    score2.append(random.randint(0, 200))

plt.rc('font', family='Nanum Gothic')
plt.title('키와 몸무게의 상관관계')
plt.scatter(height1, weight1, s=score1, color='crimson', alpha=0.7, label='그룹1')
plt.scatter(height2, weight2, s=score2, color='indigo', alpha=0.7, label='그룹2')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.legend()
plt.show()
```

```python
y = [0, 1, 2, 3, 4, 5, 6, 7]
weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]

for i in range(len(weekend)):
    weekday[i] = -weekday[i]

plt.rc('font', family='Nanum Gothic')
plt.title('주중/주말 자동차 통행량 비교')
plt.barh(y, weekday, label='주중 통행량', color='crimson')
plt.barh(y, weekend, label='주말 통행량', color='royalblue')
plt.legend()
plt.show()
```

```python
weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]

plt.title('주중/주말 자동차 통행량 비교')
plt.boxplot([weekday, weekend], labels=['주중 통행량', '주말 통행량'])
plt.show()
```
